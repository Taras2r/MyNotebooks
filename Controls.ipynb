- Use **NbConverter** extension to convert a Jupyter notebook to *html* hiding code and certain outputs


**NbConverter** [documentation](https://nbconvert.readthedocs.io/en/latest/removing_cells.html)

Learned commands those are usefull to convert a Jupyter notebook to html:
- to remove as sell input so as sell output:
```
jupyter nbconvert --to html --TagRemovePreprocessor.enabled=True --TagRemovePreprocessor.remove_cell_tags hide_cell Controls.ipynb
``` 
- to leave cell output but remove the input:
```
jupyter nbconvert --to html --TagRemovePreprocessor.enabled=True --TagRemovePreprocessor.remove_input_tags hide_cell Controls.ipynb
```

````python
import numpy as np
import matplotlib.pyplot as plt
import control as ct

t = np.linspace(1, 50, 1000) # time vector
G = ct.tf( [1, 1, 1], [1, 1, 1, 1]) # transfer function, Laplace re-presentation
````

````python
t1, y1 = ct.impulse_response(G, t)
t2, y2 = ct.step_response(G, t)
ct.sisotool(G)
````

````python
from IPython.core.display import HTML


HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
""")


fig = plt.figure(dpi=90)
plt.subplot(2, 1, 1)
plt.title("Impulse Response")
plt.grid()
plt.plot(t1, y1)
plt.subplot(2, 1, 2)
plt.title("Step Response")
plt.plot(t2, y2)
fig.tight_layout(pad=2.0)
plt.grid()
plt.show()
````
